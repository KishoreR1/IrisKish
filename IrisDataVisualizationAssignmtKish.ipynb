#Data visualization is the graphical representation of information and data. It uses visual elements like charts, graphs, and maps to help people understand complex data easily.

Data Visualization is Important in **Exploratory Data Analysis(EDA)** because it
simplifies Complex Data: Humans process visuals faster than raw numbers or tables.

Reveals Patterns & Trends: Helps identify relationships, trends, and outliers that may not be obvious from data alone.

Validates Data Quality: Visuals can uncover errors, missing data, or inconsistencies.

Guides Hypothesis Formation: Helps analysts decide what to investigate further.

Communicates Insights: Facilitates sharing findings clearly with stakeholders.



#Key Principles of Effective Data Visualization

**Clarity:** Make visuals easy to understand without ambiguity.

**Accuracy:** Represent data truthfully without distortion.

**Simplicity:** Avoid clutter; focus on key messages.

**Consistency:** Use consistent colors, scales, and formats.

**Relevance:** Choose the right chart type for the data and question.

**Interactivity:** (When possible) allows users to explore data themselves.



#Common Types of Visualizations in EDA

**Histogram:**	Understand distribution of numeric data

**Scatter Plot:**	Explore relationships between two variables

**Box Plot:**	Summarize distribution & detect outliers

**Heatmap:**	Visualize correlation or matrix-like data

**Violin Plot:**	Show distribution shape and density

**Bar Chart:**	Compare categorical data

**Line Chart:**	Show trends over time

#How Data Visualization Fits Into EDA Workflow

**Initial Data Check:** Use plots to verify data types, completeness, and distribution.

**Univariate Analysis:** Analyze individual variables using histograms or boxplots.

**Bivariate/Multivariate Analysis:** Explore relationships via scatterplots, heatmaps, and pairplots.

**Identify Issues:** Spot missing data, outliers, and anomalies.

**Feature Engineering:** Based on visual insights, create or transform features.

**Modeling Guidance:** Use visuals to choose relevant features and understand model results.

```python
import pandas as pd

iris = pd.read_csv('Iris.csv')

# Preview the first 5 rows
iris.head()
```

#1. What is the distribution of sepal lengths and widths among the Iris species?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.histplot(data=iris, x='SepalLengthCm', hue='Species', kde=True, element='step')
plt.title('Distribution of Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.show()
```

The distribution of sepal length shows that Iris Setosa tends to have shorter sepal lengths, mostly clustered around 4.5 to 5.5 cm

Iris Versicolor and Iris Virginica have generally longer sepals, with Versicolor mostly between 5.5 to 7 cm, and Virginica ranging up to 7.5 cm.

The distributions for Versicolor and Virginica overlap somewhat, indicating similar sepal lengths, but Setosa is distinctly separate with shorter sepals.

#Above is depiction of 'sepal length' by species

#Now we will see 'sepal width' by species

```python
plt.figure(figsize=(10,5))
sns.histplot(data=iris, x='SepalWidthCm', hue='Species', kde=True, element='step')
plt.title('Distribution of Sepal Width by Species')
plt.xlabel('Sepal Width (cm)')
plt.show()
```

The distribution of sepal width shows that Iris Setosa tends to have wider sepals, mostly clustered around 3.0 to 4.5 cm.
In contrast, Iris Versicolor and Iris Virginica generally have narrower sepals, with Versicolor mostly between 2.0 to 3.5 cm and Virginica showing a similar range overlapping with Versicolor.
The distributions of Versicolor and Virginica overlap quite a bit, indicating similar sepal widths, but Setosa stands out distinctly with consistently wider sepals.

#Visualize Petal Length Variation
Well use a boxplot to clearly see the distribution, median, and spread of petal lengths by species.

Run this code:

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Species', y='PetalLengthCm', data=iris)
plt.title('Petal Length Variation Across Iris Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

The above boxplot shows the following details:
**Iris Setosa:** has the shortest petal lengths, with a median around 1.5 cm and very little variation. A few outliers are present slightly above the typical range (~1.9 cm).

**Iris Versicolor:** Petal lengths are moderate, longer than setosa but shorter than virginica. Median petal length is about 4.3 cm.
The range spans roughly 3.0 to 5.1 cm.
There is one outlier on the lower side (~3.0 cm).

**Iris Virginica:** Has the longest petal lengths.
Median petal length is close to 5.5 cm.
The range is wider, from about 4.5 to 6.9 cm.
No obvious outliers

##Visualize with a scatter plot colored by species

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=iris, palette='Set1')
plt.title('Relationship Between Sepal Length and Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

#The scatter plot shows a clear positive relationship between sepal length and petal length  as sepal length increases, petal length also increases.
The species form distinct clusters: Iris Setosa has generally smaller values for both, while Versicolor and Virginica have progressively larger values.
This indicates that sepal and petal lengths together can help distinguish species.

```python
sns.pairplot(iris, hue='Species', height=2.5, palette='Set1')
plt.suptitle('Pairwise Relationships Between Iris Features', y=1.02)
plt.show()
```

#The pairplot reveals that petal length and petal width best separate the three species, with distinct clusters visible.
Sepal length and width show some overlap between species but still provide useful information.
Features like petal length and petal width have strong positive correlation, especially in Versicolor and Virginica species.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', style='Species', s=100, data=iris, palette='Set1')
plt.title('Distinguishing Iris Species Using Sepal and Petal Lengths')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

The scatter plot clearly separates Iris Setosa from the other two species.
Versicolor and Virginica overlap somewhat but tend to cluster separately.
Sepal and petal lengths combined are effective in distinguishing Iris species.

The final analysis of sepal and petal measurements shows that Iris species can be effectively distinguished based on these features.

Specifically:
Iris Setosa is clearly separable from the other species, as it consistently exhibits smaller sepal and petal lengths and widths.

#*Features like petal length and petal width provide the strongest separation among species, with relatively little overlap.

#*Combining multiple measurements in scatter plots or pairwise relationships reveals clear groupings, supporting the use of these features in classification tasks.

#*Overall, sepal and petal dimensions are highly informative and sufficient to differentiate the three Iris species with good accuracy.
