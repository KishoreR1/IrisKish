#Is there a correlation between BMI and Glucose levels?

Visualize with Scatter Plot and Correlation Coefficient

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='BMI', y='Glucose', data=df)
plt.title('Scatter Plot of BMI vs Glucose Levels')
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.show()
```

#Calculate Pearson Correlation

```python
correlation = df['BMI'].corr(df['Glucose'])
print(f"Pearson correlation between BMI and Glucose: {correlation:.2f}")
```

#Correlation Between BMI and Glucose Levels

The Pearson correlation coefficient is 0.22, which indicates a weak positive correlation between BMI and Glucose.

This means that, generally, as BMI increases, Glucose levels tend to increase slightly, but the relationship is not very strong.

The weak correlation suggests that while BMI may influence glucose, many other factors also affect glucose levels in diabetic and non-diabetic individuals.

In the scatter plot above, you can see a slight upward trend but with a lot of scatter — indicating variability.

#Observation:
There is a modest positive association between BMI and glucose, implying that higher body mass index can be somewhat related to higher blood sugar levels, but it’s not a definitive predictor on its own.

#Create age bins and plot proportion of diabetics per group

```python
# Create age groups (bins)
bins = [20, 30, 40, 50, 60, 70, 80, 90]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Calculate proportion of diabetics in each age group
age_group_diabetes = df.groupby('AgeGroup')['Outcome'].mean()

# Plot
plt.figure(figsize=(10,6))
age_group_diabetes.plot(kind='bar')
plt.title('Proportion of Diabetic Patients by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Proportion Diabetic')
plt.show()
```

#Diabetes Proportion by Age Group:
The proportion of diabetic patients tends to increase as age groups get older.

Younger groups (e.g., 20-29 and 30-39) have the lowest proportion of diabetes cases.

Middle age groups (40-49, 50-59) show a noticeable rise in diabetes prevalence.

The highest proportions are observed in the oldest age groups (60-69, 70-79, 80-89), indicating that diabetes is more common among older adults.

This trend aligns with medical understanding that risk for type 2 diabetes increases with age due to factors like reduced insulin sensitivity and cumulative lifestyle impacts.

The bar heights reflect the relative risk — as age increases, the chance of being diabetic also increases significantly.

#Observation Statement:
Diabetes prevalence clearly increases with age. The data shows that older adults have a significantly higher proportion of diabetes compared to younger individuals. This emphasizes the importance of monitoring and preventive care especially in later life stages.

#Visualize insulin levels across age groups using boxplots

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='AgeGroup', y='Insulin', data=df)
plt.title('Insulin Levels Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Insulin Level')
plt.show()
```

Insulin levels tend to vary widely within each age group, as seen by the spread of the boxes and presence of outliers.

The median insulin level (middle line of each box) remains relatively stable across younger and middle-aged groups (e.g., 20-29 up to 50-59).

For the older age groups (60 and above), there may be a slight increase or greater variability in insulin levels, suggesting more irregular insulin responses among older adults.

Many age groups show outliers, indicating individuals with unusually high or low insulin levels, which may correspond to diabetic or pre-diabetic conditions.

Overall, while the central tendency doesn’t shift dramatically, the variability and extremes in insulin levels increase in older age groups, reflecting diverse metabolic health status.
